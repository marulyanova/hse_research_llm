```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
for file in os.listdir("logs"):
    if file.endswith(".csv") and "tuning_mlp_dim" in file:
        name = file.replace(".csv", "")
        df = pd.read_csv(f"logs/{file}")
        plt.plot(df["epoch"], df["reward"], label=name.replace("tuning_mlp_dim_", "").replace("vanilla", "").replace("_", " ").strip())
plt.legend()
plt.title("Average Reward per Hidden_Dim")
plt.xlabel("Epoch")
plt.ylabel("Reward")
plt.show()
```

```python
for file in os.listdir("logs"):
    if file.endswith(".csv") and "tuning_methods_entropy_reg_coef" in file:
        name = file.replace(".csv", "")
        df = pd.read_csv(f"logs/{file}")
        plt.plot(df["epoch"], df["reward"], label=name.replace("tuning_methods_entropy_reg_coef_", "").replace("_", " ").strip())
plt.legend()
plt.title("Average Reward per Method & Entropy Regularization Coefficient")
plt.xlabel("Epoch")
plt.ylabel("Reward")
plt.show()
```
